```python
str_dataset = [[s.replace(" ", "") for s in sublist] for sublist in rows]

import re

def convert_string_to_float(s):
    try:
        # Attempt to directly convert to float
        return float(s)
    except ValueError:
        # Check for a custom scientific notation-like format (e.g., "x.xxe+xx")
        match = re.match(r"^(\d+\.\d+)e\+(\d+)$", s)
        if match:
            base, exponent = match.groups()
            float_value = float(base) * (10 ** int(exponent))
            return int(float_value)
        else:
            # If none of the above, return a placeholder
            return "no data"


# convert all elements to float, handling special formatted strings
dataset = [[convert_string_to_float(s) for s in sublist] for sublist in str_dataset]

# returning the converted list of lists
header
dataset
```
