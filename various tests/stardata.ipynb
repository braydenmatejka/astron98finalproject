```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
stars_df = pd.read_csv("stardata.txt")
stars_df
```

```python
x = stars_df.iloc[:,3]
y = stars_df.iloc[:,7]
color = stars_df.iloc[:,6]

plt.figure(figsize=(11,9))
plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Blue-Violet Color Index")
plt.ylabel("Log of Luminosity (solLum)")
# plt.gca().invert_yaxis()

plt.scatter(x, y, c=color, cmap="inferno", s=0.5)
plt.colorbar(label="Log of Effective Temp (K)")
```
